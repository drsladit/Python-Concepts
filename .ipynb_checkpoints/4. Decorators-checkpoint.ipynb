```python
# Decorator - Suppose we have a normal function f1 and then we want to add some extra functionality or code before f1.
# Then instead of deleting/rewriting the function definition. We can use decorator concept. 
# This also helps in toggle between normal and new code
# Decorator concept work on higher order function concept - where we pass the funtion and return the function as well.
# We use @decorator function name to toggle this functionality
```

```python
def dec_func(func):
    def dec_inner_fun():
        print("We want to print this decorator function")
        return func()
    
    return dec_inner_fun
```

```python
def display():
    print("Print from display function")
```

```python
dec_display = dec_func(display)
```

```python
dec_display()
```

```python
def dec_func1(func):
    def dec_inner_fun1():
        print("We want to print this decorator function")
        return func()
    
    return dec_inner_fun1
```

```python
#@dec_func1
def display1():
    print("Print from display function")
```
