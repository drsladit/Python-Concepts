```python
# Creating a Class
class Employee:
    
    # initialising or constructor method - It is also a dunder method
    def __init__(self, fname, lname, pay):
        # attributes of instance
        self.fname = fname
        self.lname = lname
        self.pay   = pay
        self.email = fname + "." + lname + "@company.com"
    
    # methods
    def fullname(self):
        fullname = self.fname + " " + self.lname
        return fullname
```

```python
# Instance creation
emp1 = Employee("Aditya", "Kothapalli", 90000)
```

```python
emp2 = Employee("Kavyasree", "Kothapalli", 90000)
```

```python
print(emp1)
print(emp2)
# We can see below that emp1 and emp2 are 2 objects/instances created at different address
```
