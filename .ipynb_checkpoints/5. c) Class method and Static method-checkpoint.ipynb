```python
# class Method: 
# Class method takes "cls" argument as defalut argument
# We need to use decorate the regular method with decorator to make it class method @classmethod
```

```python
class Employee:
    # Class Attributes/variables
    raise_amount = 1.04 # Use case 1
    no_of_emps = 0 # Use case 2
    
    # Initializing Class
    def __init__(self, fname, lname, pay):
        # Instance Attibutes/variables - values
        self.fname = fname
        self.lname = lname
        self.pay   = pay
        self.email = fname + "." + lname + "@company.com"
        
        Employee.no_of_emps += 1
    
    # Regular Methods - actions
    def fullname(self):
        return self.fname + " " + self.lname
    
    def apply_pay_raise(self):
        return self.pay*employee.raise_amount # (OR) self.pay*self.raise_amount are same.
    
    # Class method/actions
    @classmethod
    def set_raise_amt(cls, amount):
        cls.raise_amount = amount
 
```

```python
# Instance creation
emp1 = Employee("Aditya", "Kothapalli", 90000)
emp2 = Employee("Kavyasree", "Kothapalli", 90000)
```
