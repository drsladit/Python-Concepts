```python
# First class function - 
# If we can assign function object to a variable, pass the function as argument and can also be able to Return function 
# Then we can say that the lauguage has fisrtclass functions
```

```python
def square(x):
    return x*x
```

```python
f = square # Assigning function object as variable
```

```python
square
```

```python
f
```

```python
del(square)
```

```python
f
```

```python
x = map(f, [1,2,3]) # Passing function as Argument
```

```python
for n in x:
    print(n)
```

```python
# Closures - A closures closes the variable or state of the function
# Closures take advantage of higher order functions and return an inner function that remebers 
# and has access to variables local to the scope in which they were created.
```

```python
# Example 1: A simple function in a function definition where we RETURN FUNCTION
def outer_fn():
    message = "hi"
    
    def inner_fn():
        print(message)
    
    return inner_fn()
```

```python
outer_fn()
```

```python
#Example 2: A simple function in a function definition where we RETURN FUNCTION OBJECT

def outer_fn1():
    message = "hi"
    
    def inner_fn1():
        print(message)
    
    return inner_fn1
```

```python
outer_fn1()
```

```python
myfunc = outer_fn1()
```

```python
myfunc
```

```python
myfunc()
```

```python
#Example 3: A simple function in a function definition where we Return function OBJECT and then pass the argument
def outer_fn2(msg):
    message = msg
    
    def inner_fn2():
        print(message)
    
    return inner_fn2
```

```python
hi_fnc = outer_fn2("hi")
hello_func = outer_fn2("Hello")
```

```python
hi_fnc
```

```python
hello_func
```

```python
hi_fnc()
```

```python
hello_func()
```
