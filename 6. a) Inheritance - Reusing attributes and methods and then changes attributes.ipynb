```python
# Inheritance:
# By inheriting a class we can use all the attributes and methods from parent class
# Method resolution order: It is order in which python checks for attributes and methods in some order from child, parent etc
```

```python
class Employee:
    
    raise_amount = 1.04
    
    def __init__(self, fname, lname, pay):
        self.fname = fname
        self.lname = lname
        self.pay   = pay
        self.email = fname + "." + lname + "@company.com"
        

    def fullname(self):
        return self.fname + " " + self.lname
    
    def apply_pay_raise(self):
        self.pay = self.pay*self.raise_amount # there will be difference in using Employee.raise_amount and self.raise_amount
```

```python
# by just inheriting the class we can see below we can use - __init__, attibutes, methods from parent class
#class Developer(Employee):
#    pass
```

```python
# We can just change the raise amount/attibiute values for it to completelyreflect across for all developers
class Developer(Employee):
    raise_amount = 1.10
```

```python
dev_1 = Developer("aditya", "kothapalli", 93000)
dev_2 = Developer("Kavya", "Alluri", 50000)
```

```python
print(dev_1.email)
print(dev_1.raise_amount)
print(dev_2.fullname())
```

```python
# Observe this case

print(dev_1.pay)
dev_1.apply_pay_raise()
print(dev_1.pay)
```
