```python
#Classes - Logically group data and functions and make us easy to reuse and easy to build i.e add or remove.
```

```python
class employee:
    # Class Attributes/variables
    emp_no = 0
    raise_amount = 1.04
    
    # Initializing Class
    def __init__(self, fname, lname, pay):
        # Instance Attibutes/variables
        self.fname = fname
        self.lname = lname
        self.pay   = pay
        self.email = fname + "." + lname + "@company.com"
        
        employee.emp_no += 1
    
        # methods
    def fullname(self):
        return self.fname + " " + self.lname
    
    def apply_pay_raise(self):
        return self.pay*employee.raise_amount # OR self.pay*self.raise_amount
        
```

```python
# Instance creation
emp1 = employee("aditya", "kothapalli", 30000)
emp2 = employee("kavyasree", "kothapalli", 25000)
```
