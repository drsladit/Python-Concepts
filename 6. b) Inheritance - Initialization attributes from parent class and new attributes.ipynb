```python
class Employee:
    
    raise_amount = 1.04
    
    def __init__(self, fname, lname, pay):
        self.fname = fname
        self.lname = lname
        self.pay   = pay
        self.email = fname + "." + lname + "@company.com"
        

    def fullname(self):
        return self.fname + " " + self.lname
    
    def apply_pay_raise(self):
        self.pay = self.pay*self.raise_amount # there will be difference in using Employee.raise_amount and self.raise_amount
```

```python
class Developer(Employee):
    raise_amount = 1.10
    def __init__(self, first, last, pay, prog_lang):
        super().__init__(first, last, pay) # (or) Employee.__init__(self, first, last, pay)
        self.prog_lang = prog_lang
```

```python
dev_1 = Developer("aditya", "kothapalli", 93000, "python")
dev_2 = Developer("Kavya", "Alluri", 50000, "Bid manager")
```
