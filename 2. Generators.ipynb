```python
# Generators are like normal function except that instead of returning a single value.
# They create iterator function with help of yield and generates sequence of values.
# Generators are more memory efficient than normal function or list comprehension when we have large amount of data.
```

```python
# fibonacci series

def fib(f):
    result = []
    a,b = 0,1
    if f <=0:
        print("Invalid")
    elif f == 1:
        print(a)
    else:
        while a<f:
            result.append(a)
            b,a = a+b,b
    return result
```

```python
fib(500)
```

```python
# fibonacci series

def fibg(f):
    #result = []
    a,b = 0,1
    if f <=0:
        print("Invalid")
    elif f == 1:
        print(a)
    else:
        while a<f:
            yield a
            a,b = b,a+b
```

```python
fn = fibg(50)
```

```python
for x in fn:
    print(x)
```
