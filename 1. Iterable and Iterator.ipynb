```python
# Iterable - List is an iterable but not iterator. To convert iterable to iterator use __iter__ Special method
# Iterator - It is an object with a state which remembers where it is stopped during iteration.
```

```python
# Uses of Iterators: Itertors build memory efficient code by just remembering the state of the iterable object. 
# This is achieving by fetching one value at a time instead of holding all values in list ot tuple.
# By this way we will not store complete data in storage. It will just store current state.
```

```python
# To fetch the results one by one we can use iter method with next method/for loop or Generators with next or for loop
```

```python
nums = [1,2,3]
```

```python
type(nums)
```

```python
next(nums)
```

```python
#dir(nums)  # nums has __iter__ method. Hence iterable
```

```python
i_nums = iter(nums)
```

```python
type(i_nums)
```

```python
#dir(i_nums) # i_nums has both __iter__ and __next__ methods . Hence iterator.
```

```python
next(i_nums)
```
